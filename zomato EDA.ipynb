```python
df_c=pd.read_excel('country-code.xlsx')
df_c.head()
```

```python
f_df=pd.merge(df,df_c,on='Country Code',how='left')
```

```python
f_df.head(20)
```

```python
f_df.columns
```

```python
c_names=f_df.Country.value_counts().index
```

```python
c_values=f_df.Country.value_counts().values
```

```python
#significance of 1.2f
```

```python
plt.pie(c_values[:3],labels=c_names[:3],autopct='%1.2f%%')
```

```python
#{ } for rating
```

```python
ratings=f_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
#hue rcParamas
#why figure.figsize
#palette=
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,4)
sns.barplot(x='Aggregate rating',y="Rating count",hue="Rating color",data=ratings)
```
